### Filtering Data in Pandas

```python
# Boolean Indexing
filtered_df = df[df['sales'] > 5000]
filtered_df
```

```python
#to get unqiue values of a column
df['category'].unique()
```

```python
# value_counts() will give all the counts of records for each category
df['category'].value_counts()
```

```python
# value_counts() will give all the counts of records for each city for each category
df[['category', 'city']].value_counts()
```

```python
# multiple conditions
fc = (df.category == 'Furniture') | (df.category == 'Technology')
df[fc]
```
